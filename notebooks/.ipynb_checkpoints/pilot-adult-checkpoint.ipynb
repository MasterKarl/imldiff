# Adult Data Set
Predict whether income exceeds \\$50K/yr based on census data.
([UCI Repository](https://archive.ics.uci.edu/ml/datasets/adult))

We will compare a **logistic regression model (A)** and an **XGBoost logistic regression model (B)**.

Sources:
- [Explaining a linear logistic regression model](https://shap.readthedocs.io/en/latest/example_notebooks/general/Explainable%20AI%20with%20Shapley%20Values.html#Explaining-a-linear-logistic-regression-model)
- [Explaining an XGBoost logistic regression model](https://shap.readthedocs.io/en/latest/example_notebooks/general/Explainable%20AI%20with%20Shapley%20Values.html#Explaining-an-XGBoost-logistic-regression-model)

```python
import shap
import sklearn
import xgboost
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import warnings
from datetime import datetime

# deactivate warnings that clutter the notebook
# warnings.filterwarnings('ignore')

# activate matplotlib interactive style
%matplotlib widget
```

```python
X,y = shap.datasets.adult()
```

```python
# a simple linear logistic model
model_a = sklearn.linear_model.LogisticRegression(max_iter=10000)
model_a.fit(X, y)

def model_a_proba(x):
    return model_a.predict_proba(x)[:,1]
def model_a_odds_ratios(x):
    probabilities = model_a_proba(x)
    return probabilities / (1 - probabilities)
def model_a_log_odds(x):
    p = model_a.predict_log_proba(x)
    return p[:,1] - p[:,0]
```

```python
# train XGBoost model
model_b = xgboost.XGBClassifier(nestimators=100, max_depth=2).fit(X, y)

def model_b_proba(x):
    return model_b.predict_proba(x)[:,1]
def model_b_odds_ratios(x):
    probabilities = model_b_proba(x)
    return probabilities / (1 - probabilities)
def model_b_log_odds(x):
    return np.log(model_b_odds_ratios(x))
```

Let's build the difference models, that enable us to explain the differences directly

## Partial dependence plots
Attention: the predicted probabilities of a logistic regression model are not well suited to be explained.

```python
sample_ind = 18
# TODO subplots does not work with partial_dependence_plot
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", model_a_proba, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for probabilities of A')
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", model_a_odds_ratios, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for odds ratios of A')
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", model_a_log_odds, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for log-odds of A')
```

```python
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", model_b_proba, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for probabilities of B')
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", model_b_odds_ratios, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for odds ratios of B')
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", model_b_log_odds, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for log-odds of B')
```

```python
fig,ax = shap.partial_dependence_plot(
    "Age", model_a_proba, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for probabilities of A')
fig,ax = shap.partial_dependence_plot(
    "Age", model_b_proba, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for probabilities of B')
```

Now let's simply create the differences of the model outputs in probability space and log of odds space

```python
def diff_proba(x):
    return model_b_proba(x) - model_a_proba(x)
def diff_odds_ratios(x):
    return model_b_odds_ratios(x) - model_a_odds_ratios(x)
def diff_log_odds(x):
    return model_b_log_odds(x) - model_a_log_odds(x)
```

```python
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", diff_proba, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for differences of probabilities')
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", diff_odds_ratios, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for differences of odds ratios')
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", diff_log_odds, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for differences of log-odds')
```

```python
fig,ax = shap.partial_dependence_plot(
    "Age", diff_proba, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for differences of probabilities')
fig,ax = shap.partial_dependence_plot(
    "Age", diff_odds_ratios, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for differences of odds ratios')
fig,ax = shap.partial_dependence_plot(
    "Age", diff_log_odds, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for differences of log-odds')
```

```python
def simil_proba(x):
    return model_b_proba(x) * model_a_proba(x)
def simil_odds_ratios(x):
    return model_b_odds_ratios(x) * model_a_odds_ratios(x)
def simil_log_odds(x):
    return model_b_log_odds(x) * model_a_log_odds(x)
```

```python
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", simil_proba, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for similarities of probabilities')
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", simil_odds_ratios, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for similarities of odds ratios')
fig,ax = shap.partial_dependence_plot(
    "Capital Gain", simil_log_odds, X, model_expected_value=True,
    feature_expected_value=True, show=False, ice=False
)
ax.set_title('Partial dependence for similarities of log-odds')
```

```python
explainer_model_a_proba = shap.Explainer(model_a_proba, X)
explainer_model_a_odds_ratios = shap.Explainer(model_a_odds_ratios, X)
explainer_model_a_log_odds = shap.Explainer(model_a_log_odds, X)

explainer_model_b_proba = shap.Explainer(model_b_proba, X)
explainer_model_b_odds_ratios = shap.Explainer(model_b_odds_ratios, X)
explainer_model_b_log_odds = shap.Explainer(model_b_log_odds, X)

explainer_diff_proba = shap.Explainer(diff_proba, X)
explainer_diff_odds_ratios = shap.Explainer(diff_odds_ratios, X)
explainer_diff_log_odds = shap.Explainer(diff_log_odds, X)

explainer_simil_proba = shap.Explainer(simil_proba, X)
explainer_simil_odds_ratios = shap.Explainer(simil_odds_ratios, X)
explainer_simil_log_odds = shap.Explainer(simil_log_odds, X)
```

```python
from sklearn.utils import resample
display_data = shap.datasets.adult(display=True)[0].values
X_small, y_small, display_data_small = resample(X, y, display_data, n_samples=int(len(X)/10), replace=False, stratify=y, random_state=0)
```

```python
started = datetime.now()
explainer_simil_proba = shap.Explainer(simil_proba, X_small)
shap_values_simil_proba = explainer_simil_proba(X_small)
shap_values_simil_proba.display_data = display_data
datetime.now() - started
```

```python
# TODO divide by zero encountered in log
# TODO invalid value encountered in double_scalars
started = datetime.now()
shap_values_model_a_proba = explainer_model_a_proba(X)
shap_values_model_a_odds_ratios = explainer_model_a_odds_ratios(X)
shap_values_model_a_log_odds = explainer_model_a_log_odds(X)

shap_values_model_b_proba = explainer_model_b_proba(X)
shap_values_model_b_odds_ratios = explainer_model_b_odds_ratios(X)
shap_values_model_b_log_odds = explainer_model_b_log_odds(X)

shap_values_diff_proba = explainer_diff_proba(X)
shap_values_diff_odds_ratios = explainer_diff_odds_ratios(X)
shap_values_diff_log_odds = explainer_diff_log_odds(X)

shap_values_simil_proba = explainer_simil_proba(X)
shap_values_simil_odds_ratios = explainer_simil_odds_ratios(X)
shap_values_simil_log_odds = explainer_simil_log_odds(X)
datetime.now() - started
```

## Scatter plots for the Shapeley values of _Age_
They are generalizations of the partial dependence plots.

In the graph for model B, we instruct SHAP to pick the feature that has the biggest interaction effects with _Age_ and to colour the observations according to its' shapeley values.

```python
fig1, (ax1, ax2) = plt.subplots(ncols=1, nrows=2, constrained_layout=True)
shap.plots.scatter(shap_values_a[:,"Age"], color=shap_values_a, ax=ax1)
shap.plots.scatter(shap_values_b[:,"Age"], color=shap_values_b, ax=ax2)
```

```python
fig1, (ax1, ax2) = plt.subplots(ncols=1, nrows=2, constrained_layout=True)
shap.plots.scatter(shap_values_simil_proba[:,"Age"], color=shap_values_simil_proba, ax=ax1)
shap.plots.scatter(shap_values_simil_proba[:,"Capital Gain"], color=shap_values_simil_proba, ax=ax2)
```

## Bar plots
By default a SHAP bar plot will take the mean absolute value of each feature over all the instances (rows) of the dataset.

```python
fig1, (ax1) = plt.subplots(ncols=1, nrows=1, constrained_layout=True)
shap.plots.bar(shap_values_simil_proba, max_display=12)
```

```python
shap.plots.bar(shap_values_a, max_display=12)
shap.plots.bar(shap_values_b, max_display=12)
```

But the mean absolute value is not the only way to create a global measure of feature importance, we can use any number of transforms. Here we show how using the max absolute value highights the Capital Gain and Capital Loss features, since they have infrewuent but high magnitude effects.

```python
shap.plots.bar(shap_values_a.abs.max(0), max_display=12)
shap.plots.bar(shap_values_b.abs.max(0), max_display=12)
```

```python
shap.plots.bar(shap_values_a.abs.min(0), max_display=12)
shap.plots.bar(shap_values_b.abs.min(0), max_display=12)
```

```python
sex = ["Women" if shap_values_a[i,"Sex"].data == 0 else "Men" for i in range(shap_values_a.shape[0])]
shap.plots.bar(shap_values_a.cohorts(sex).abs.mean(0), max_display=12)
shap.plots.bar(shap_values_b.cohorts(sex).abs.mean(0), max_display=12)
```

## Beeswarm plots: Shapeley value distribution per feature
If we are willing to deal with a bit more complexity we can use a beeswarm plot to summarize the entire distribution of SHAP values for each feature.

```python
fig1, (ax1) = plt.subplots(ncols=1, nrows=1, constrained_layout=True)
shap.plots.beeswarm(shap_values_simil_proba)
```

```python
shap.plots.beeswarm(shap_values_a)
shap.plots.beeswarm(shap_values_b)
```

By taking the absolute value and using a solid color we get a compromise between the complexity of the bar plot and the full beeswarm plot. Note that the bar plots above are just summary statistics from the values shown in the beeswarm plots below.

```python
shap.plots.beeswarm(shap_values_a.abs, color="shap_red")
shap.plots.beeswarm(shap_values_b.abs, color="shap_red")
```
