[SHAP docs:](https://shap.readthedocs.io/en/latest/example_notebooks/overviews/An%20introduction%20to%20explainable%20AI%20with%20Shapley%20values.html)
> Note that explaining the probability of a linear logistic regression model is not linear in the inputs.

```python
from data_loader import load_data
from models import LogisticProbabilityModel, XGBoostProbabilityModel
from util import printmd
import shap
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
```

```python
X, y, display_data = load_data()
linear_model = LogisticProbabilityModel()
linear_model.load_or_train(X, y)
tree_based_model = XGBoostProbabilityModel()
tree_based_model.load_or_train(X, y)
```

```python
models = [linear_model, tree_based_model]
```

```python
def partial_dependence_plot(feature_name):
    printmd(f'### {feature_name}')
    for model in models:
        printmd(f'#### {type(model).__name__}')
        shap.partial_dependence_plot(
            feature_name, model.predict, X, model_expected_value=True,
            feature_expected_value=True, show=False, ice=False)
        plt.show()
```

```python
partial_dependence_plot('Age')
```
