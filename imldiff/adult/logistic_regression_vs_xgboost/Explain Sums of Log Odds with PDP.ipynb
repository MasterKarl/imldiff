# Explain Sums of Log Odds with PDP

```python
from imldiff.adult.loader import load_data
from imldiff.adult.logistic_regression_vs_xgboost.models import LogisticLogOddsModel, XGBoostLogOddsModel
from imldiff.comparers import SumComparer
from imldiff.util import printmd
import shap
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
```

```python
X, y, display_data = load_data()
feature_names = list(X.columns)
linear_model = LogisticLogOddsModel()
linear_model.load_or_train(X, y)
tree_based_model = XGBoostLogOddsModel()
tree_based_model.load_or_train(X, y)
```

```python
comparer = SumComparer(linear_model, tree_based_model)
```

```python
def partial_dependence_plot(feature_name):
    printmd(f'### {feature_name}')
    shap.partial_dependence_plot(
        feature_name, comparer.predict, X, model_expected_value=True,
        feature_expected_value=True, show=False, ice=False)
    plt.show()
```

```python
partial_dependence_plot('Age')
```
